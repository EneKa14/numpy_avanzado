### Utiliza Python y Numpy para realizar las siguientes operaciones:
1. Dado el arreglo `a = np.array([1, 2, 3, 4])`, determina si ninguno de los elementos es cero.
2. Dado el arreglo `a = np.array([0, 0, 2, 0])`, determina si contiene algún elemento que no sea cero.
3. Analiza el arreglo `a = np.array([1, 2, np.nan, np.inf])` y determina si contiene valores que sean infinitos o valores no numéricos (NaN).
4. Dado el arreglo `a = np.array([1, 2, np.inf])`, determina si contiene valores infinitos.
5. Dado el arreglo `a = np.array([1, np.nan, 3])`, determina si contiene valores NaN.
6. Dados los arreglos `a = np.array([1, 3, 5])` y `b = np.array([2, 3, 4])`, realiza una comparación elemento a elemento contemplando mayor, mayor o igual, menor y menor o igual.
7. Crea un arreglo que contenga los valores 1, 7, 13 y 105, y determina la cantidad de memoria que ocupa el arreglo.
8. Crea un arreglo de 10 ceros; uno de 10 unos; y uno de 10 cincos.
9. Crea un arreglo que contenga todos los números enteros entre 30 y 70.
10. Crea un arreglo que contenga los números pares entre 30 y 70.
11. Crea una matriz identidad de 3x3.
12. Genera un número aleatorio entre 0 y 1.
13. Crea un arreglo de 15 números aleatorios tomados de una distribución normal estándar.
14. Crea un arreglo con valores entre 15 y 55, e imprime todos los valores excepto el primero y el último.
15. Crea un vector (arreglo) de longitud 10 con valores equitativamente distribuidos entre 5 y 50.
16. Crea un vector con valores de 0 a 20, y cambia el signo de los números de 9 a 15.
17. Convierte el arreglo `a = np.array([1, 2, 3])` a una lista de Python.
18. Convierte la lista `l = [1, 2, 3]` a un arreglo de Numpy.
19. Dada la matriz `m = np.array([[1, 2], [3, 4]])`, verifica qué números son pares y almacena los resultados en otra matriz.
20. Dado el arreglo `m = np.array([[1, 2, 3], [4, 5, 6]])`, encuentra el número de filas y columnas de la matriz.

```python
import numpy as np
```

```python
# ejercicio 1
a1 =np.array([1,2,3,4])
print("1. ¿Ninguno es cero?:" , np.all(a1 != 0))
```

```python
# ejercicio 2
a2 = np.array ([0, 0, 2, 0])
print("2. ¿Contiene algún no-cero?:" , np.any(a2 != 0))
```

```python
# ejercicio 3
a3 = np.array([1,2, np.nan, np.inf])
print("3. ¿Tiene NaN?:" , np.any(np.isnan(a3)))
print(" ¿Tiene Infinito?:" , np.any(np.isinf(a3)))
```

```python
# ejercicio 4
a4 = np.array([1,2, np.inf])
print(" ¿Tiene infinitos?:" , np.any(np.isinf(a4)))
```

```python
# ejercicio 5
a5 = np.array([1, np.nan, 3])
print("5. ¿Tiene NaN?:" , np.any(np.isnan(a5)))
```

```python
# ejercicio 6
a6 = np.array ([1,3,5])
b6 = np.array([2,3,4])
print ("6. Comparaciones:")
print(" Mayor que:", a6 > b6)
print("Mayor o igual:" , a6 >= b6)
print ("Menor que:" , a6<b6)
print (" Menor o igual: " , a6 <= b6)
```

```python
# ejercicio 7
a7 = np.array ([1,7,13,105])
print (f"\n7. Memoria del arreglo: {a7.nbytes} bytes")
```

```python
# ejercicio 8
ceros = np.zeros(10)
unos = np.ones(10)
cincos = np.full(10,5)
print (f"\n8. Ceros: {ceros}")
print(f" Unos: {unos}")
print(f" Cincos: {cincos}")
```

```python
# ejercicio 9
enteros = np.arange(30, 71)
print(f"\n9. Rango 30-70: {enteros}")
```

```python
# ejercicio 10
pares =np.arange(30, 71, 2)
print (f"\n10. Pares 30-70: {pares}")
```

```python
# ejercicio 11
identidad = np.eye(3)
print(f"\n11. Matriz Identidad:\n{identidad}")
```

```python
# ejercicio 12
aleatorio =np.random.rand()
print(f"12. Número aleatorio único: {aleatorio}")
```

```python
# ejercicio 13
normal_std = np.random.randn(15)
print (f"\n13. 15 números aleatorios (Normal):\n{normal_std}")
```

```python
# ejercicio 14
rango_completo = np.arange (15,56)
sin_extremos =rango_completo [1:-1]
print (f"\n14. Rango 15-55 sin extremos: \n{sin_extremos}")
```

```python
# ejercicio 15
equidistantes = np.linspace(5,50,10)
print(f"\n15. Vector linspace (5 a 50):\n {equidistantes}")
```

```python
# ejercicio 16
vec_20 = np.arange(21)
vec_20[(vec_20 >= 9) & (vec_20 <=15)] *= -1
print(f"\n16. Cambio de signo (9 al 15):\n {vec_20}")
```

```python
# ejercicio 17
arreglo_a = np.array([1,2,3])
lista_nueva = arreglo_a.tolist()
print(f"\n17. Convertido a lista: {lista_nueva} (Tipo: {type(lista_nueva)})")
```

```python
# ejercicio 18
lista_l = [1,2,3]
arreglo_nuevo = np.array(lista_l)
print(f"18. Convertido a arreglo: {arreglo_nuevo} (Tipo: {type (arreglo_nuevo)})")
```

```python
# ejercicio 19
m_pares = np.array([[1,2], [3,4]])
solo_pares = m_pares[m_pares % 2 ==0]
print(f"\n19. Matriz original:\n{m_pares}")
print (f" Valores pares encontrados: {solo_pares}")
```

```python
# ejercicio 20
m_dim = np.array([[1,2,3], [4,5, 6]])
filas, columnas = m_dim.shape
print(f"\n20. La matriz tiene {filas} filas y {columnas} columnas.")
```
